### Soal no. 2

```python
# 1. Lakukan tahapan pada soal 1 dengan menggunakan tensorflow

import tensorflow as tf

x_tf = tf.Variable(28, dtype=tf.int16)
x_tf
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable(30, dtype=tf.int16)
```

```python
y_tf.shape
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
x_tf_float = tf.Variable(28., dtype=tf.float16)
x_tf_float
```

```python
tf_sum.numpy()
```

```python
tf_float_sum = x_tf_float + tf.cast(y_tf, tf.float16)
tf_float_sum
```

```python
x_tf_float.shape
```
